# ANN Model

```python
# Artificial Neural Network
# Part 1 - Data Preprocessing
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Churn_Modeling.csv')
X = dataset.iloc[:, 3:12].values
y = dataset.iloc[:, 12].values
```

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder
labelencoder_X = LabelEncoder()
X[:, 1] = labelencoder_X.fit_transform(X[:, 1])
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Part 2 - Now let's make the ANN!
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initialising the ANN
classifier = Sequential()
# Adding the input layer and the first hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 9))
```

```python
# Adding the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.fit(X_train, y_train, batch_size = 5, epochs = 10)
```

```python
classifier.fit(X_train, y_train, batch_size = 10, epochs = 10)
```

```python
classifier.fit(X_train, y_train, batch_size = 20, epochs = 10)
```

```python
classifier.fit(X_train, y_train, batch_size = 100, epochs = 20)
```

```python
classifier.fit(X_train, y_train, batch_size =200, epochs = 20)
```

#### Highest accuracy: 0.8337 (batch:200, epoch 20) &  (batch:100, epoch 20)

```python
classifier.fit(X_train, y_train, batch_size =200, epochs = 50)
```

###### Highest accuracy -  accuracy: 0.8339 (Batch Size= 200, epoch=50)

```python
# Part 3 - Making predictions and evaluating the model
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
ctr=0
for i in range(0,df.shape[0],1):
    if df.loc[i].Gender == 'Male':
        df.loc[ctr,['Gender']] = 1
    else:
        df.loc[ctr,['Gender']] = 0
    ctr = ctr+1

df
```

```python
df.Gender=df['Gender'].astype('int64')
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(X_train,y_train)
y_predict=reg.predict(X_test)
```

```python
reg.score(X_test,y_test)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_regression
testScores = []
trainScores = []
```

```python
for i in range(1, len(df.columns)):
    fs = SelectKBest(score_func = mutual_info_regression, k=i)
    fs.fit(X_train, y_train)
    X_train_fs = fs.transform(X_train)
    X_test_fs = fs.transform(X_test)
    reg.fit(X_train_fs, y_train)
    
    testScore = reg.score(X_test_fs, y_test)
    trainScore = reg.score(X_train_fs, y_train)
    testScores.append(testScore)
    trainScores.append(trainScore)
```

```python
df = pd.DataFrame(data={'testScores':testScores,
                        'trainScores':trainScores})
df.reset_index(inplace=True)
df.rename(columns={'index':'numFeatures'}, inplace=True)
df.numFeatures = df.numFeatures + 1
df.plot(x='numFeatures', y=['testScores','trainScores'])
```

```python
print(df.testScores.idxmax(), df.testScores.max())
```
